## 6) New types at runtime(dynamically) 

üìè This can create new types in a fully dynamic way. It‚Äôs the same as creating a class but something new !

```python
# Python program to create new type object
  
# Creates a new type object
NewType = type("NewType", (object, ), {"attr": "hello newtype"})
New = NewType()
  
# Print the type of object
print(type(New))
  
# Print the attribute of object
print(New.attr)
```

```python
# OR

# Creates a class
class NewType:
    attr = "hello newtype"
    
# Initialize an object
New = NewType()
  
# Print the type of object
print(type(New))
  
# Print the attribute of object
print(New.attr)
```

## 7) sys.exit()

We can cause the program to terminate by calling the sys.exit() function. Since this function is in the sys module, firstly, the sys module should be imported.

```python
# sys.exit()

# Cause the program to terminate by calling the sys.exit() function.

# This will import sys module
import sys
  
while True:
    print("Type 'exit' to exit")
    response = input()
    if response.lower() == "exit":
        print("Exiting the program")
        sys.exit()
    print("You typed", response)
```

## 8) urllib

üí® Urllib module is the URL handling module for python. It is used to fetch URLs (Uniform Resource Locators). It uses the &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;urlopen function and is able to fetch URLs using a variety of different protocols.

üí® Urllib is a package that collects several modules for working with URLs, such as:

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;‚óæ urllib.request for opening and reading.<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;‚óæ urllib.parse for parsing URLs<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;‚óæ urllib.error for the exceptions raised<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;‚óæ urllib.robotparser for parsing robot.txt files

```python
# This will import urlopen class from urllib module

from urllib.request import urlopen

page = urlopen("https://www.amazon.in/")
print(page.headers)
```

### You can also see the coding of the website by using read() function 

```python
# This will import urlopen class from urllib module
  
from urllib.request import urlopen
page=urlopen("https://www.amazon.in/")
  
# Fetches the code of the web page

content = page.read()
  
print(content)
```

## 9) Turtle

‚úî Turtle is a Python module to draw. <br>
‚úî It has a huge application and a number of methods which you can learn about in [here](https://docs.python.org/2/library/turtle.html)!. <br>
‚úî But with just a few basics, pretty cool stuff can be done.

```python
# Turtle

# This will import turtle module
import turtle
  
myTurtle = turtle.Turtle()
myWin = turtle.Screen()
  
# Turtle to draw a spiral
def drawSpiral(myTurtle, linelen):
    myTurtle.forward(linelen)
    myTurtle.right(90)
    drawSpiral(myTurtle, linelen-10)
    
drawSpiral(myTurtle, 80)
myWin.exitonclick()
```
